### ༗ PLOT USING DATAFRAME ༗

```python
import pandas as pd
import matplotlib.pyplot as plt
a=pd.DataFrame({"A":[1,2,3,4,5],"B":[5,4,3,2,1]})
plt.plot(a["A"],a["B"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
a=[10,15,35,45,55]
b=[20,25,40,50,50]               ###DEFAULT linewidth 1 pixel###
d=[10,20,30,40,50]
c=pd.DataFrame([a,b,d])
print(c)
c.plot(kind='line',linewidth=1,xlabel="Age",ylabel="Height(in cm)",markersize=2,color=['r','m','b','y','g'])
plt.show()
```
