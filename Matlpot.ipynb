### ✯ PLOT USING DATAFRAME ✯

## ✯ LINE PLOT✯

```python
#  xlabel()       to write label name for x-axis         # 
#  ylabel()       to write label name for y-axis         #
#  title()        to write title for the plot            #
#  legend()       to show legends                        # 
#  xticks()       To set ticks for x axis                #
#  yticks()       To set ticks for y axis                #
#  grid()         To show gridlines                      #
#  show()         to view the plot                       #
#  savefig()      To save the plot as .png or .pdf at    #
#                 the desired location                   #
##########Syntax: plot(x, y, color, marker, markersize, linewidth, linestyle, label)#######
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
a=[1,2,3]
b=[1,2,3]
c=[1,4,8]                                    #default 0.8#
plt.plot(a,b,color='r',marker="8",markersize=6,linewidth=2,linestyle="--")
plt.plot(c,b,color='b',marker="d",markersize=6,linewidth=2.7,linestyle="-.")
plt.title("NEW Plot",color='c',fontsize='30',fontname='Arial')
plt.xlabel("numbers ->",fontsize =20,color='b',fontname="Comic Sans MS")
plt.ylabel("numbers ->",fontsize =20,color='r',fontname="Comic Sans MS")
plt.legend("ab") # or label="a",label="b" in the respective plot syntax#
plt.grid()
                #NO TICKS#
plt.savefig("JOJO")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
a=[]
b=[]
x=0
while x<=30:
    x+=0.001
    a+=[x]
    b=b+[math.cos(x)]          
c=pd.DataFrame(b,a)
c.plot()
plt.show()
```
